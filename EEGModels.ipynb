```python
import pandas as pd
import numpy as np
```

This is the start of a script I'm working on for a Kaggle competition. The idea is you get a series of EEGs and you need to predict a few set of tasks. You are only allowed to use information that occurred before the task started. There are 6 tasks: HandStart, FirstDigitTouch, BothStartLoadPhase, LiftOff, Replace, BothReleased. So far, I've 1) loaded the data sets for one subject into a data frame and plotted the EEGs for a subset of the data where HandStart is *not* occurring; 2) passed the data frame to an array. This way I can change the labelling for the subject's time to be an integer value instead of a string (function removeStr). Then I can run through the array and add all the start and end times to the list (function getStartAndEndTimes). Next, I plan to use these times so I can isolate pieces of the EEG before the event and compare these pieces for different occurrences of the events. Then the pieces which are similar enough serve as the predictor for an event.

```python
df_events = pd.read_csv('subj1_series8_events.csv',header=0)
```

```python
df_data = pd.read_csv('subj1_series8_data.csv',header=0)
```

```python
df_events = df_events.drop(['id'],axis=1)
```

```python
result = pd.concat([df_events,df_data],axis=1)
```

```python
def removeStr(string,start):
    string = string[start:]
    return int(string)
```

```python
result_arr = result.values
```

```python
for i in xrange(0,117333):
    result_arr[i,6]=removeStr(result_arr[i,6],14)
```

```python
handStart_range=[]
firstDigitTouch_range=[]
bothStartLoadPhase_range=[]
liftOff_range=[]
replace_range=[]
bothRelease_range=[]
```

```python
def getStartAndEndTimes(array,column,store):
    store.append(0)
    for i in xrange(0,117332):
        handStart_currentValue = array[i,column]
        handStart_nextValue = array[i+1,column]
        if handStart_currentValue == 0 and handStart_nextValue == 1:
            store.append(array[i,6])
```

```python
getStartAndEndTimes(result_arr,0,handStart_range)
getStartAndEndTimes(result_arr,1,firstDigitTouch_range)
getStartAndEndTimes(result_arr,2,bothStartLoadPhase_range)
getStartAndEndTimes(result_arr,3,liftOff_range)
getStartAndEndTimes(result_arr,4,replace_range)
getStartAndEndTimes(result_arr,5,bothRelease_range)
```

```python
def plotSimilar(array,sensor):
    i=0
    while i<(len(array)-2):
        result.loc[array[i]:array[i+1],:].plot(x=['id'],y=[sensor],figsize=(10,10))
        i=i+2
```
